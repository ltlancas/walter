# Resolved Stellar Populations with Roman

In this notebook we will walk through the formalism laid out in the associated paper (put link here) for calculating the expected number of stars resolved in a given observation. This code is intended to make it easier for those who would like to apply this formalism for planning observing strategies of resolved stellar populations. Though this code and formalism is applicable generally, all examples given here will assume the user is observing with the future Nancy Grace Roman Space Telescope (Roman). We will note all places where code would need to be edited to apply this code to a different observing instrument.

The plots made in this notebook rely on the pre-calculation of a few quantites that depend on the stellar population that is being considered for observation. Each of these quantities depends on an stellar isochrone package, we use two separate examples, below `ISO` is meant as a placeholder for `mist` or `padova`, either of which can be used, more on this below. If one was using an observing instrument other than Roman, the following quatities would need to be re-computed. These precalculated quantities and the associated code used to calculate them are
1. $f_{\rm det}(B, \tau, {\rm [Fe/H]},B_{\rm cut})$ The number fraction of a population's stars that are observed, which is a function of the band being observed in, $B$, the population parameters (just age $\tau$ and metallicity ${\rm [Fe/H]}$ in our case), and the limiting magnitude that is reached in the observation $B_{\rm cut}$. This is calculated in the notebook `fdet_calc_ISO.ipynb` and the corresponding output file is `fdet_ISO.npy`
2. $\left\langle L_{B,\tau,{\rm [Fe/H]}}\right\rangle$ The average luminosity of a given stellar population in a given band. This is calculated in `Lavg_calc_ISO.py` and the corresponding output file is `lum_ISO.npy`
3. $\left\langle L_{B,\tau,{\rm [Fe/H]}}^2\right\rangle_{L<L_{\rm lim}}$ the second moment of luminosities less than some limiting luminosity. This is used to calculate the crowding limit of an observation based on the reasoning given in Olsen et al. 2003. This quantity is calculated in `l2_calc_ISO.ipynb` and the corresponding output file is `l2_ISO.npy` These values are (by default) calculated for the same range of limiting absolute magnitudes (equivalent to the limiting luminsoity) as for $f_{\rm det}$.

The code for the calculation of these quantities is provided with this repository, with implementations both in Python (which is slower) and Cython (which should be reasonably fast). For convenience, we provide these values pre-calculated for two sets of isochrones from stellar evolution codes. The first code we use is the Module for Experiments in Stellar Astrophysics (MESA) Isochrones and Stellar Tracks (MIST) code (`ISO -> mist` above) and the second is the Padova Stellar Ischrone library (`ISO -> padova` above). The results from either code can be used to make figures in this notebook by changing only a few lines, which we will explicitly point out, but the default setup is for MIST.

We will proceed in making plots of these quantities in the order given above and focus on making simple plots first followed by making plots more similar to those found in our paper.

# Plotting the Fraction of a Population that is Detected

We will have to do this separately for each band pass.

```python
# first calculate the distance modulus
mu_R = ps_detect_5slim[0] - Mags + 1.25*np.log10(texp/1e3)
mu_Z = ps_detect_5slim[1] - Mags + 1.25*np.log10(texp/1e3)
mu_Y = ps_detect_5slim[2] - Mags + 1.25*np.log10(texp/1e3)
mu_J = ps_detect_5slim[3] - Mags + 1.25*np.log10(texp/1e3)
mu_W = ps_detect_5slim[4] - Mags + 1.25*np.log10(texp/1e3)
mu_H = ps_detect_5slim[5] - Mags + 1.25*np.log10(texp/1e3)
mu_F = ps_detect_5slim[6] - Mags + 1.25*np.log10(texp/1e3)
## only for Padova
#mu_K = ps_detect_5slim[7] - Mags + 1.25*np.log10(texp/1e3)

# then the distance in Mpc
d_R = 10**((mu_R-25)/5)
d_Z = 10**((mu_Z-25)/5)
d_Y = 10**((mu_Y-25)/5)
d_J = 10**((mu_J-25)/5)
d_W = 10**((mu_W-25)/5)
d_H = 10**((mu_H-25)/5)
d_F = 10**((mu_F-25)/5)
## only for Padova
#d_K = 10**((mu_K-25)/5)

# collect distances in to a single object
ds = np.array([d_R,d_Z,d_Y,d_J,d_W,d_H,d_F])
## for Padova
#ds = np.array([d_R,d_Z,d_Y,d_J,d_W,d_H,d_F,d_K])
```

It is clear from the plot above that there are certain points in distance where $f_{\rm det}$ decreases drastically. As indicated in Figure 1 of the paper, these transitions correspond to certain stellar evolutionary stages. For example, the last major drop is indicated by the Tip of the Red Giant Branch (TRGB). Locating these evolutionary stages will depend on the particular stellar evolutionary code that was used for calculating the various pre-calculated parameters that we loaded above. In the MIST package we use their Equivalent Evolutionary Phases (EEPs) to define certain parts of the stellar evolutionary sequence, we then identify the brightest members of this evolutionary phase. An example of how we do this for the $F158$ band is given in the function below.

```python
def get_evol_mags_F158(a,met,path_to_isos):
    # arguments ar age "a" in log_10(age/years) so 9.0 == billion year old pop
    # metalliccity as "met" in log_10 with reference 0 at solar
    # and a string specifying the path to where the isochrones are stored "path_to_isos" 

    # we start by loading the specified MIST isochrone based on metallicity
    if met>-0.1:
        (eep,age,imass,T,L) = np.loadtxt("%s/mist_fehp%1.2f.iso.cmd"%(path_to_isos,abs(met)),usecols=[0,1,2,4,6]).T
        (F158) = np.loadtxt("%s/mist_fehp%1.2f.iso.cmd"%(path_to_isos,abs(met)),usecols=[14]).T
    else:
        (eep,age,imass,T,L) = np.loadtxt("%s/mist_fehm%1.2f.iso.cmd"%(path_to_isos,abs(met)),usecols=[0,1,2,4,6]).T
        (F158) = np.loadtxt("%s/mist_fehm%1.2f.iso.cmd"%(path_to_isos,abs(met)),usecols=[14]).T

    # We then restrict to the age under consideration
    sids = np.intersect1d(np.where(age>a-0.01),np.where(age<a+0.01))
    # This selects the Main sequence EEP <= 495, the brightest part of this will be the MSTO
    ids_ms = np.intersect1d(sids,np.where(eep<=495))
    # This selects the TRGB 560 <= EEP <= 580
    ids_trgb = np.intersect1d(sids,np.where(np.logical_and(eep>=560,eep<=580)))
    # This selects the Horizontal Branch/Red Clump 630 <= EEP <= 640
    ids_hb = np.intersect1d(sids,np.where(np.logical_and(eep>=630,eep<=640)))
    # We then select the brightest magnitude in each evolutionary stage
    m_trgb = min(F158[ids_trgb])
    m_hb = min(F158[ids_hb])
    m_ms = min(F158[ids_ms])
    return (m_trgb,m_hb,m_ms)
```

# The Stellar Density for a given Population

Here we will describe how to make plots similar to Figure 2 of our paper, which shows the total number density of stars (both observed and unobserved) for a given population. The only parameters required for making this plot are the surface brightness of the population being considered ($Sigma$) and the average luminosity of a star in this population $\left\langle L_{B,\tau, {\rm [Fe/H]}}\right\rangle$. We first determine the range of distances we want to look at.

```python
# set the distance range that you want to look over
(minD,maxD) = (1.,25.)
nDs = 200
drange = np.linspace(minD,maxD,nDs)
# then give distancce modulus
mus = 5*np.log10(drange*1e5)
```

We are then ready to make some plots showing the total stellar density of a population given it's surface brightness and distance. We'll start by making a plot for a single stellar population, one age, one metallicity, and at a given  surface brightness.

```python
meti = 5
agej = 10
Sigma = 32 # in mag/arcsec^2

print("We are considering a population of [Fe/H] = %1.2f, Age = %1.2f Gyrs, and  Sigma = %d mag/arcsec^2"%(metlist[meti],10**(agelist[agej]-9),Sigma))
```

```python
# plot n_tot as a function of distance for each filter
for filtk in range(nfilt):
    ntot = 10**((Sigma-mus)/-2.5)/lum[meti,agej,filtk]
    plt.plot(drange,ntot,c=plt.cm.viridis((1.*filtk)/(nfilt-1)),label=filter_names[filtk])
idmid = nDs//2
nref = ntot[idmid]

# Some limits and labeling of the axes
plt.xlim(1,25)
plt.yscale("log")
plt.ylabel(r"$n_{\rm tot}\, [{\rm arcsec}^{-2}]$")
plt.xlabel(r"$d \, [{\rm Mpc}]$")
# put in a legend
plt.legend()
# Putting in the labels for what population we're considering
plt.text(drange[idmid],nref/10,r"$\Sigma$ = %1.0f mag/arcsec$^2$"%(Sigma),fontsize=18)
plt.text(drange[idmid],nref/30,r"[Fe/H] = %1.2f"%(metlist[meti]),fontsize=18)
plt.text(drange[idmid],nref/90,r"$\tau$ = %1.2f Gyrs"%(10**(agelist[agej]-9)),fontsize=18)
plt.show()
```

Next we'll make a multi-panel plot that shows this quantity for a range of populations (metallicities, ages, and surface brightnesses). We'll start by specificing the different values we want to plot.

```python
# metal indices to plot
met_toplot = [2,6,10]
nmtp = len(met_toplot)
# Surface brightness values to plot
Sig_toplot = [20,24,28,32]
# Ages to plot, we'll just do two
agej1 = 22
agej2 = 13
```

We then plot the results for each of the above for 2 filters $F062$ and $F184$. We'll vary the color of the plotted line for varying metallicity and the style of line for the two different ages. We won't change anything for the different surface brightnesses as they should be far enough apart to tell the difference anyway.  This should be similar to Figure 2 of the paper.

```python
ax1 = plt.subplot(221)
filtj = 0
for (mi,meti) in enumerate(met_toplot):
    for Sig in Sig_toplot:
        ax1.plot(drange,10**((Sig-mus)/-2.5)/lum[meti,agej1,filtj],c=plt.cm.coolwarm((0.5+mi*1.)/nmtp),lw=0.75)
for (mi,meti) in enumerate(met_toplot):
    for Sig in Sig_toplot:
        ax1.plot(drange,10**((Sig-mus)/-2.5)/lum[meti,agej2,filtj],'--',c=plt.cm.coolwarm((0.5+mi*1.)/nmtp),lw=0.75)
ax1.set_title(r"%s filter"%(filter_names[filtj]))
ax1.set_yscale("log")
ax1.set_ylim(1e-2,5e6)
ax1.set_xlim(1,25)
ax1.set_ylabel(r"$n_{\rm tot} \, [{\rm arcsec}^{-2}]$")

ax2 = plt.subplot(222)
filtj = 5
for (mi,meti) in enumerate(met_toplot):
    for Sig in Sig_toplot:
        ax2.plot(drange,10**((Sig-mus)/-2.5)/lum[meti,agej1,filtj],c=plt.cm.coolwarm((0.5+mi*1.)/nmtp),lw=0.75)
for (mi,meti) in enumerate(met_toplot):
    for Sig in Sig_toplot:
        ax2.plot(drange,10**((Sig-mus)/-2.5)/lum[meti,agej2,filtj],'--',c=plt.cm.coolwarm((0.5+mi*1.)/nmtp),lw=0.75)
ax2.set_title(r"%s filter"%(filter_names[filtj]))
ax2.set_yscale("log")
ax2.set_ylim(1e-2,5e6)
ax2.set_xlim(1,25)


plt.gcf().set_size_inches((2.*3.37, 2*3.37*.8))
plt.tight_layout()
plt.show()
```

# Crowding Limits

Finally, we want to show how to calculate the crowding limit of an observation, this involves the final quantity that we have pre-calculated: $\left\langle L_{B,\tau, {\rm [Fe/H]}}^2\right\rangle_{L<L_{\rm lim}}$. As with the other plots we will begin by making a simple, single age and metallicity stellar population at a fixed surface brightness.

```python
# Again, we specify the details of the population under consideration
meti = 5
agej = 15
Sigma = 32 # in mag/arcsec^2

print("We are considering a population of [Fe/H] = %1.2f, Age = %1.2f Gyrs, and  Sigma = %d mag/arcsec^2"%(metlist[meti],10**(agelist[agej]-9),Sigma))
```

The next function will help us by interpolating in order to evaluate the last term in Equation 6 of our paper.

```python
def get_olsen_term(sigmaB,A,Bcut,Lavg,l2_Mags,l2):
    # interpolate the l2 values calculated for a range of limiting 
    # absolute magnitudes on to the limiting absolute magnitudes
    # calculated here
    L2_interp = np.interp(Bcut[::-1],l2_Mags[::-1],l2[::-1])[::-1]
    # calculate the last term in Equation 8 in our paper
    return 2.5*np.log10(((sigmaB/1.086)**2)*(Lavg)/(A*L2_interp))
```

These next parameters are from the Olsen '03 paper and we assume some simple values,

```python
sigm = 0.01 # photometric precision of measurement in magnitudes
ares = (0.1)**2 # psf area in square arceseconds (assumed top hat) this should really vary with filter
```

```python
# plot the surface brightness brighter than which we cannot accurately
# measure the magnitudes of the stars that we should be able to detect
for filtk in range(nfilt):
    mus = 5*np.log10(ds[filtk]*1e5)
    Mcut = ps_detect_5slim[filtk] - mus
    ols_fac = get_olsen_term(sigm,ares,Mcut,lum[meti,agej,filtk],Mags,L2[meti,agej,filtk])
    L2_interp = np.interp(Mcut[::-1],Mags[::-1],L2[meti,agej,filtk][::-1])[::-1]
    lso =  2*Mcut + mus - ols_fac
    
    plt.plot(ds[filtk],lso,c=plt.cm.viridis((1.*filtk)/(nfilt-1)),label=filter_names[filtk])

# Some limits and labeling of the axes
plt.xlim(1,25)
plt.ylim(32,28)
plt.ylabel(r"$\Sigma \, [{\rm mag}\, {\rm arcsec}^{-2}]$")
plt.xlabel(r"$d \, [{\rm Mpc}]$")
# put in a legend
plt.legend()
plt.show()
```

# $n_{\rm det}$ at $\Sigma_{\rm lim}$

Okay, one last thing. We now want to take a look at what the number density of detected stars would be if we observed at the limiting surface brightness for a given distance, exposure time, band, etc. This is really the quantity that we care about most, what is the best we can do? 

```python
# plot the surface brightness brighter than which we cannot accurately
# measure the magnitudes of the stars that we should be able to detect
for filtk in range(nfilt):
    mus = 5*np.log10(ds[filtk]*1e5)
    Mcut = ps_detect_5slim[filtk] - mus
    ols_fac = get_olsen_term(sigm,ares,Mcut,lum[meti,agej,filtk],Mags,L2[meti,agej,filtk])
    L2_interp = np.interp(Mcut[::-1],Mags[::-1],L2[meti,agej,filtk][::-1])[::-1]
    # this is our limiting surface brightness
    lso =  2*Mcut + mus - ols_fac
    # this is the corresponding total number of stars
    ntot_here = 10**((Sig-mus)/-2.5)/lum[meti,agej1,filtj]
    # and fraction that are detected as a function of distance
    fdet_here = fdet[meti,agej1,:,filtj+1]
    
    plt.plot(ds[filtk],ntot_here*fdet_here,c=plt.cm.viridis((1.*filtk)/(nfilt-1)),label=filter_names[filtk])

# Some limits and labeling of the axes
plt.xlim(1,25)
plt.ylim(2e-4,1e-2)
plt.yscale("log")
plt.ylabel(r"$n_{\rm det}\, [{\rm arcsec}^{-2}]$")
plt.xlabel(r"$d \, [{\rm Mpc}]$")
# put in a legend
plt.legend()
plt.show()
```

The more interesting question is probably, what is the maximum exposure time when I start getting less return of stars on my investment of telescope time. This code can be modified to tackle that problem as well. We'll specify a fixed surface brightness, population age/metallicity, and distance. This will give us a plot similar to Figure 4 in the paper.

```python
# Again, we specify the details of the population under consideration
meti = 10
agej = 22
Sigma = 32 # in mag/arcsec^2
d = 10 # in Mpc
# distance modulus
mu = 25 + 5*np.log10(d)

print("We are considering a population of [Fe/H] = %1.2f, Age = %1.2f Gyrs"%(metlist[meti],10**(agelist[agej]-9)))
print("and Sigma = %d mag/arcsec^2 at a distance of d = %d Mpc"%(Sigma,d))
```

```python
# a list of exposure times
expts = np.logspace(1,6,100)
# corresponding limiting absolute magnitudes
Rlim = ps_detect_5slim[0] - mu + 1.25*np.log10(expts/1e3)
Zlim = ps_detect_5slim[1] - mu + 1.25*np.log10(expts/1e3)
Ylim = ps_detect_5slim[2] - mu + 1.25*np.log10(expts/1e3)
Jlim = ps_detect_5slim[3] - mu + 1.25*np.log10(expts/1e3)
Wlim = ps_detect_5slim[4] - mu + 1.25*np.log10(expts/1e3)
Hlim = ps_detect_5slim[5] - mu + 1.25*np.log10(expts/1e3)
Flim = ps_detect_5slim[6] - mu + 1.25*np.log10(expts/1e3)
## only for Padova
#Klim = ps_detect_5slim[7] - mu + 1.25*np.log10(expts/1e3)
# put them together in an array (need to add Klim for Padova)
Mlims = np.array([Rlim,Zlim,Ylim,Jlim,Wlim,Hlim,Flim])

## determine the actual number density of stars, which is determined
## by the above. This varies from band to band because a fixed surface 
## brightness has a different meaning in each band.
ntot_R = 10**((Sigma-mu)/-2.5)/lum[meti,agej,0]
ntot_Z = 10**((Sigma-mu)/-2.5)/lum[meti,agej,1]
ntot_Y = 10**((Sigma-mu)/-2.5)/lum[meti,agej,2]
ntot_J = 10**((Sigma-mu)/-2.5)/lum[meti,agej,3]
ntot_W = 10**((Sigma-mu)/-2.5)/lum[meti,agej,4]
ntot_H = 10**((Sigma-mu)/-2.5)/lum[meti,agej,5]
ntot_F = 10**((Sigma-mu)/-2.5)/lum[meti,agej,6]
#ntot_K = 10**((Sigma-mu)/-2.5)/lum[meti,agej,7]

## put them together in an array (need to add ntot_K for Padova)
ntots = np.array([ntot_R,ntot_Z,ntot_Y,ntot_J,ntot_W,ntot_H,ntot_F])

for filtk in range(nfilt):
    Mlim = Mlims[filtk]

    ols_fac = get_olsen_term(sigm,ares,Mlim,lum[meti,agej,filtk],Mags,L2[meti,agej,filtk])
    # this is our limiting surface brightness for
    # crowding as a function of exposure time
    lso =  2*Mlim + mu - ols_fac
    # this is the corresponding total number of stars
    # at that limiting surface brightness
    ntot_lim = 10**((lso-mu)/-2.5)/lum[meti,agej,filtj]
    # and fraction that are detected as a function of exposure time
    fdets_here = np.interp(Mlim,fdet[meti,agej,:,0][::-1],fdet[meti,agej,:,filtk+1][::-1])

    ## this is the number density of detected stars if we ignore the effects of crowding
    ndet_no_sb_lim = ntots[filtk]*fdets_here
    ## this is the numbder density of detected stars "if" we were observing a population
    ## which had a surface brightness equal to our limiting surface brightness due to crowding
    ndet_at_sb_lim = ntot_lim*fdets_here
    ## We then want to plot the ndet curve only up until the point at which it becomes 
    ## crowding limited, so we select all points where ndet_no_sb_lim < ndet_at_sb_lim
    sids = np.where(ndet_no_sb_lim - ndet_at_sb_lim < 0)
    plt.plot(expts[sids],ndet_no_sb_lim[sids],c=plt.cm.viridis((1.*filtk)/(nfilt-1)),label=filter_names[filtk])
    if (expts[sids][-1] < expts[-1]):
        plt.plot([expts[sids][-1]],[ndet_no_sb_lim[sids][-1]],'o',c=plt.cm.viridis((1.*filtk)/(nfilt-1)))

# you may need to adjust these limits if you adjust the parameters you are plotting for
plt.ylim(1e-6,1e-1)
plt.xlim(min(expts),max(expts))
plt.legend()
plt.yscale("log")
plt.xscale("log")
plt.xlabel(r"$t_{\rm exp}\, [{\rm seconds}]$")
plt.ylabel(r"$n_{\rm det}\, [{\rm arcsec}^{-2}]$")
plt.show()
```

The above plot is very helpful in deciding what exposure time we want to choose for a given observation. We see that the most drastic benefits come after a $10^3$ second exposure, with steady increasing returns up to about $2\times 10^4 $ seconds (which would be quite long anyway!).
